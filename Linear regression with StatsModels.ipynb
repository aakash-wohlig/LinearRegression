```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
data = pd.read_csv('satgpa.csv')
```

```python
data
```

```python
data.describe()
```

```python
y = data['fy_gpa']
x1 = data['sat_sum']
```

```python
plt.scatter(x1,y)
plt.xlabel('SAT Score')
plt.ylabel('GPA')
plt.show()
```

```python
x = sm.add_constant(x1)
results = sm.OLS(y, x).fit()
results.summary()
```

```python
plt.scatter(x1, y)
yhat = 0.0239*x1 + 0.0019
fig = plt.plot(x1, yhat, lw=4, c='orange', label='regression line')
plt.xlabel('GPA', fontsize=20)
plt.ylabel('SAT Score', fontsize=20)
plt.show()
```

```python
plt.scatter(x1, y)
yhat = 0.0239*x1 + 0
fig = plt.plot(x1, yhat, lw=4, c='orange', label='regression line')
plt.xlabel('GPA', fontsize=20)
plt.ylabel('SAT Score', fontsize=20)
plt.xlim(0)
plt.ylim(0)
plt.show()
```

```python
plt.scatter(x1, y)
yhat = 0*x1 + 0.0019 
fig = plt.plot(x1, yhat, lw=4, c='red', label='regression line')
plt.xlabel('GPA', fontsize=20)
plt.ylabel('SAT Score', fontsize=20)
plt.show()
```
